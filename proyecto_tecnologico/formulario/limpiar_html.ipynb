```python
from bs4 import BeautifulSoup

def html_to_text(file_path):
    with open(file_path, "r", encoding="utf-8") as f:
        html = f.read()

    soup = BeautifulSoup(html, "lxml")

    # Eliminar scripts y estilos
    for tag in soup(["script", "style", "noscript"]):
        tag.extract()

    # Obtener el texto limpio
    text = soup.get_text(separator=" ")

    # Limpiar espacios en blanco múltiples
    lines = (line.strip() for line in text.splitlines())
    chunks = (phrase.strip() for line in lines for phrase in line.split("  "))
    cleaned_text = "\n".join(chunk for chunk in chunks if chunk)

    return cleaned_text
```

```python
import re
import unicodedata

def normalize_text(text):
    text = text.lower()
    text = unicodedata.normalize('NFD', text)
    text = text.encode('ascii', 'ignore').decode("utf-8")
    text = re.sub(r'[^a-z0-9\s]', '', text)
    text = re.sub(r'\s+', '_', text.strip())
    return text
```

```python
name = "Quito honesto"
texto = html_to_text("./paginas/" + name + ".html")
print(texto[:1000])  # muestra los primeros 1000 caracteres

plain_name = normalize_text(name)
with open("paginas_limpias/" + plain_name + ".txt", "w", encoding="utf-8") as f:
    f.write(texto)
```
