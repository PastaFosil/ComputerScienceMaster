```python
def get_option(text):
    pattern = r'(?:^|\s|\*|\"|\(|:)([a-i])(\)|,|\.)'
    m = re.findall(pattern, str(text))
    m_set = set([i[0] for i in m])
    return ' '.join(m_set) if m else np.nan
```
