```python
import os
import unicodedata
import re
```

```python
# Carpeta donde est√°n tus archivos descargados
carpeta = "../buscador/all_pages"

def limpiar_nombre(nombre):
    """Elimina acentos y reemplaza espacios por guiones bajos en el nombre."""
    # Quitar acentos
    nombre = unicodedata.normalize("NFKD", nombre)
    nombre = nombre.encode("ascii", "ignore").decode("utf-8")
    # Reemplazar espacios por guiones bajos
    nombre = nombre.replace(" ", "_")
    # Eliminar caracteres no v√°lidos
    nombre = re.sub(r"[^\w\.-]", "", nombre)
    return nombre

# Recorrer todos los archivos en la carpeta
for nombre in os.listdir(carpeta):
    ruta_original = os.path.join(carpeta, nombre)
    if os.path.isfile(ruta_original):
        nuevo_nombre = limpiar_nombre(nombre)
        ruta_nueva = os.path.join(carpeta, nuevo_nombre)
        
        # Evitar sobrescribir archivos con el mismo nombre
        if ruta_original != ruta_nueva:
            os.rename(ruta_original, ruta_nueva)
            print(f"‚úÖ Renombrado: {nombre} ‚Üí {nuevo_nombre}")

print("\nüéâ Renombrado completado.")
```
