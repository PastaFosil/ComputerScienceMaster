```python
import os
import re
```

```python
directorio = "./respuestas_formulario_chatgpt"
archivos = os.listdir(directorio)

paginas_dict = {}
paginas_dict["qui"] = '7'
paginas_dict["con"] = '4'
paginas_dict["el_"] = '6'
paginas_dict["pol"] = '9'
paginas_dict["fic"] = '2'
paginas_dict["cor"] = '3'
paginas_dict["por"] = '5'
paginas_dict["asa"] = '8'

cols = ['pagina', 'tipo_entidad', 'origen_entidad', 'proposito',
       'informacion_presentada', 'vision', 'consecuencias', 'que_muestra',
       'intencion', 'localizacion_problema', 'involucrados', 'formalidad',
       'nivel_educacion', 'tono', 'emociones']

archivos = [f for f in archivos if f.endswith('.txt')]
print(len(archivos))
```

```python
archivos
```

```python
directorio_out = "./"
respuestas_todas = [cols[0] + ';version;' + ';'.join(cols[1:]) + '\n']
respuestas_todas_pol = [cols[0] + ';version;' + ';'.join(cols[1:]) + '\n']
for archivo in archivos:
    if paginas_dict[archivo[11:14]] == "9" and (archivo[-7] != "2"):
        incisos = [archivo[-7], archivo[-5]]
        with open(os.path.join(directorio, archivo), 'r', encoding='utf-8') as f:
            txt = f.read()

        partes = re.split(r'(?im)(?=^\s*Pregunta\s*\d+\s*:)', txt)
        for parte in partes:
            if not parte.strip():
                continue
            m = re.findall(r'[ \n]([a-i]\))', parte)
            incisos.append(', '.join(m))
            respuestas = ';'.join(incisos) + '\n'
        respuestas_todas_pol.append(respuestas)
    elif (paginas_dict[archivo[11:14]] == "9") and (archivo[-7] == "2"):
        incisos = [paginas_dict[archivo[11:14]], archivo[-5]]
        incisos_pol = [archivo[-7], archivo[-5]]
        with open(os.path.join(directorio, archivo), 'r', encoding='utf-8') as f:
            txt = f.read()

        partes = re.split(r'(?im)(?=^\s*Pregunta\s*\d+\s*:)', txt)
        for parte in partes:
            if not parte.strip():
                continue
            m = re.findall(r'[ \n]([a-i]\))', parte)
            incisos.append(', '.join(m))
            incisos_pol.append(' '.join(m))
            respuestas = ';'.join(incisos) + '\n'
            respuestas_pol = ';'.join(incisos_pol) + '\n'
        respuestas_todas.append(respuestas)
        respuestas_todas_pol.append(respuestas_pol)
    else:
        incisos = [paginas_dict[archivo[11:14]], archivo[-5]]
        with open(os.path.join(directorio, archivo), 'r', encoding='utf-8') as f:
            txt = f.read()


        partes = re.split(r'(?im)(?=^\s*Pregunta\s*\d+\s*:)', txt)
        for parte in partes:
            if not parte.strip():
                continue
            m = re.findall(r'[ \n]([a-i]\))', parte)
            incisos.append(', '.join(m))
            respuestas = ';'.join(incisos) + '\n'
        respuestas_todas.append(respuestas)

with open(os.path.join(directorio_out, directorio + "/incisos.csv"), 'w', encoding='utf-8') as f:
    f.writelines(respuestas_todas)

with open(os.path.join(directorio_out, directorio + "/incisos_pol.csv"), 'w', encoding='utf-8') as f:
    f.writelines(respuestas_todas_pol)
```
