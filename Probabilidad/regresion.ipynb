```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,20,200)


epsilon = np.random.normal(0, x)
real_beta = 2
y = real_beta*x+epsilon

beta = np.sum(y)/np.sum(x)
print(beta)

plt.scatter(x, y)
plt.plot(x, beta*x, 'r', label="Ajuste lineal")
plt.plot(x, real_beta*x, 'g', label="Real")
plt.legend()
```
