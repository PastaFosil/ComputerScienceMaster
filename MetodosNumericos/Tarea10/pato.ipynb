```python
import numpy as np
import matplotlib.pyplot as plt


# Datos obtenidos
x = [0.900000, 1.300000, 1.900000, 2.100000, 2.600000, 3.000000, 3.900000, 4.400000, 4.700000, 5.000000, 6.000000, 7.000000, 8.000000, 9.200000, 10.500000, 11.300000, 11.600000, 12.000000, 12.600000, 13.000000, 13.300000]
a = [1.300000, 1.500000, 1.850000, 2.100000, 2.600000, 2.700000, 2.400000, 2.150000, 2.050000, 2.100000, 2.250000, 2.300000, 2.250000, 1.950000, 1.400000, 0.900000, 0.700000, 0.600000, 0.500000, 0.400000, 0.250000]
b = [0.539624, 0.420752, 1.086803, 1.294942, 0.593399, -0.022191, -0.503406, -0.477075, -0.071316, 0.262340, 0.080776, 0.014558, -0.139008, -0.335834, -0.531830, -0.731178, -0.492949, -0.141335, -0.178900, -0.392775]
c = [0.000000, -0.297179, 1.407263, -0.366567, -1.036519, -0.502457, -0.032226, 0.084888, 1.267642, -0.155455, -0.026109, -0.040108, -0.113458, -0.050564, -0.100202, -0.148983, 0.943082, -0.064048, 0.001440, -0.536126]
d = [-0.247649, 0.946912, -2.956382, -0.446635, 0.445051, 0.174160, 0.078076, 1.314171, -1.581219, 0.043115, -0.004666, -0.024450, 0.017471, -0.012728, -0.020325, 1.213405, -0.839275, 0.036382, -0.447971, 0.595695]


# Función para evaluar el spline
def spline(x_val, a, b, c, d, x_i):
    return a + b*(x_val - x_i) + c*(x_val - x_i)**2 + d*(x_val - x_i)**3


# Crear la gráfica del spline
plt.figure(figsize=(20, 3))


# Evaluar el spline en cada subintervalo
for i in range(len(x) - 1):
    # Generar puntos equiespaciados entre x_i y x_{i+1}
    x_interval = np.linspace(x[i], x[i + 1], 100)
    y_interval = spline(x_interval, a[i], b[i], c[i], d[i], x[i])
    plt.plot(x_interval, y_interval, 'b')


# Graficar los puntos originales
plt.plot(x, a, 'ro', label='Puntos dados')

plt.xlabel('x')
plt.ylabel('S(x)')
plt.title('Interpolación con Spline Cúbico Natural')
plt.legend()
plt.grid()
plt.show()
```
