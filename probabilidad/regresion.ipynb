```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,20,200)
n = 5000
beta = np.empty(n)
real_beta = 2

epsilon = np.random.normal(0, x)
y = real_beta*x+epsilon
plt.scatter(x, y, c='b', s=5, label="Observaciones")
beta[0] = np.sum(y)/np.sum(x)
plt.plot(x, beta[0]*x, 'r', label="Ajuste")

for i in range(1,n):
    epsilon = np.random.normal(0, x)
    y = real_beta*x+epsilon
    plt.scatter(x, y, c='b', s=5)
    
    beta[i] = np.sum(y)/np.sum(x)

    plt.plot(x, beta[i]*x, 'r')
plt.plot(x, real_beta*x, 'g', label="Real")
plt.legend()
```

```python
h = 2*(np.percentile(beta,75)-np.percentile(beta,25))/n**(1/3)
bins = int((np.max(beta)-np.min(beta))/h)
plt.hist(beta,bins)
plt.show()
```
