```python
import numpy as np
import matplotlib.pyplot as plt


# Datos obtenidos
x1 = [1.0, 2.0, 5.0, 6.0, 7.0, 8.0, 10.0, 13.0, 17.0]
a1 = [3.0, 3.7, 3.9, 4.2, 5.7, 6.6, 7.1, 6.7, 4.5]
b1 = [1.0, 0.44681, -0.0744797, 1.01634, 1.40911, 0.54722, 0.0484602, -0.338131, 3.88549e-318, 4.04739e-320]
c1 = [-0.34681, -0.20638, 0.0326168, 1.05821, -0.665439, -0.19645, -0.0529297, -0.0759343, -0.00703287, 0.0]
d1 = [0.04681, 0.0265552, 0.341863, -0.574548, 0.156329, 0.0239201, -0.00255607, 0.00574178, 0.0, 0.0]

x2 = [17.0, 20.0, 23.0, 24.0, 25.0, 27.0, 27.7]
a2 = [4.5, 7.0, 6.1, 5.6, 5.8, 5.2, 4.1]
b2 = [3, -0.197875, -0.608501, -0.111371, 0.153987, -0.401178, 3.88549e-318, 3.95253e-323]
c2 = [-1.10071, 0.0347502, -0.171626, 0.668756, -0.403398, 0.125815, -5.26699, 1.01185e-320]
d2 = [0.126162, -0.0229307, 0.280127, -0.357385, 0.0882022, -2.568, 0, 0]

x3 = [27.7, 28.0, 29.0, 30.0]
a3 = [4.1, 4.3, 4.1, 3.0]
b3 = [0.33, 0.661386, -0.765347, 0, 0]
c3 = [2.26205, -1.15743, -0.269307, -0.465347, 3.88549e-318]
d3 = [-3.79941, 0.29604, -0.0653465, 1.01185e-320, 0]

# Función para evaluar el spline
def spline(x_val, a, b, c, d, x_i):
    return a + b*(x_val - x_i) + c*(x_val - x_i)**2 + d*(x_val - x_i)**3


# Crear la gráfica del spline
plt.figure(figsize=(20, 3))


# Evaluar el spline en cada subintervalo
for i in range(len(x1) - 1):
    # Generar puntos equiespaciados entre x_i y x_{i+1}
    x_interval = np.linspace(x1[i], x1[i + 1], 100)
    y_interval = spline(x_interval, a1[i], b1[i], c1[i], d1[i], x1[i])
    plt.plot(x_interval, y_interval, 'b')

# Graficar los puntos originales
plt.plot(x1, a1, 'ro', label='Puntos dados')

plt.xlabel('x1')
plt.ylabel('S(x1)')
plt.title('Interpolación con Spline Cúbico Natural')

for i in range(len(x2) - 1):
    # Generar puntos equiespaciados entre x_i y x_{i+1}
    x_interval = np.linspace(x2[i], x2[i + 1], 100)
    y_interval = spline(x_interval, a2[i], b2[i], c2[i], d2[i], x2[i])
    plt.plot(x_interval, y_interval, 'b')


# Graficar los puntos originales
plt.plot(x2, a2, 'ro')

for i in range(len(x3) - 1):
    # Generar puntos equiespaciados entre x_i y x_{i+1}
    x_interval = np.linspace(x3[i], x3[i + 1], 100)
    y_interval = spline(x_interval, a3[i], b3[i], c3[i], d3[i], x3[i])
    plt.plot(x_interval, y_interval, 'b')

# Graficar los puntos originales
plt.plot(x3, a3, 'ro')

plt.xlabel('x')
plt.ylabel('S(x)')
plt.title('Interpolación con Spline Cúbico Natural')
plt.legend()
plt.grid()
plt.show()
```
